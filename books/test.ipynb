```python
#importing necessary libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#reading the webpage
url = "https://books.toscrape.com/"
response = requests.get(url)
response.status_code
```

```python
#parsing the webpage(only one page)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
#extracting book details & inserting into df
book_list = []
for book in soup.find_all("li",class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
    title = book.find("h3").text
    price = book.find("p",class_="price_color").text
    rating = book.find("p",class_="star-rating")["class"][1]
    availability = book.find("p",class_="instock availability").text.split()[0]
    book = {
        "title": title,
        "price": price,
        "rating": rating,
        "availability": availability
    }
    book_list.append(book)
df_raw = pd.DataFrame(book_list)
df_raw.info()
df_raw.head()
```

```python
#price is in string format, we need to convert it to float
df_raw["price"] = df_raw["price"].str.replace("Â£", "")
df_raw.info()
df_raw.head()
df_raw["price"] = df_raw["price"].astype(float)
df_raw.info()
```

```python
#rating is in string format, we need to convert it to numeric(eg:"Three" to 3)
import helper_functions as hf
df_raw = hf.convert(df_raw, "rating")
df_raw.info()
df_raw.head()
```

```python
#availability is in string format, we need to convert it to boolean(eg:"In stock" to True)
df_raw = hf.convert_to_boolean(df_raw, "availability")
df_raw.info()
df_raw.head()
```
